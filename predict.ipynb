```python
import pickle
import numpy as np
import os

# --- Step 1: Load the Saved Model ---
file_path = 'model/house_model.pkl'

# Check if the model file exists
if not os.path.exists(file_path):
    print("Error: Model file not found! Please run the training script first.")
else:
    model = pickle.load(open(file_path, 'rb'))
    print("Model Loaded Successfully!")
    print("-" * 40)

    # --- Step 2: Take User Input ---
    print("Please enter the details of the house:")
    
    try:
        # Numerical Inputs
        area = int(input("Area (in sq ft): "))
        bedrooms = int(input("Number of Bedrooms: "))
        bathrooms = int(input("Number of Bathrooms: "))
        stories = int(input("Number of Stories: "))
        
        # Binary Inputs (1 for Yes, 0 for No)
        print("\n--- For the following, enter 1 for YES and 0 for NO ---")
        mainroad = int(input("Is it on the Main Road? (1/0): "))
        guestroom = int(input("Does it have a Guestroom? (1/0): "))
        basement = int(input("Does it have a Basement? (1/0): "))
        hotwater = int(input("Does it have Hot Water Heating? (1/0): "))
        ac = int(input("Does it have Air Conditioning? (1/0): "))
        
        # Parking (Numerical)
        parking = int(input("Number of Parking Spots (0, 1, 2...): "))
        
        # Binary Input
        prefarea = int(input("Is it in a Preferred Area? (1/0): "))
        
        # Furnishing Status
        print("\n--- Furnishing Status (Enter 1 if applicable, else 0) ---")
        semi = int(input("Is it Semi-Furnished? (1/0): "))
        unfurnished = int(input("Is it Unfurnished? (1/0): "))
        # Logic: If both are 0, the model assumes it is 'Furnished'.

        # Prepare the input array (Must match the training order)
        user_input = [[area, bedrooms, bathrooms, stories, mainroad, guestroom, 
                       basement, hotwater, ac, parking, prefarea, semi, unfurnished]]

        # --- Step 3: Make Prediction ---
        predicted_price = model.predict(user_input)
        
        # --- Step 4: Display Result ---
        print("\n" + "="*40)
        print(f"💰 Estimated House Price: ₹ {predicted_price[0]:,.2f}")
        print("="*40)

    except ValueError:
        print("\nError: Please enter valid integer numbers only!")
```
